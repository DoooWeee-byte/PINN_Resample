```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
xx = np.linspace(-40, 40, 801)
```

```python
xx.shape
```

```python
tt = np.linspace(0, 120, 1201)
```

```python
tt.shape
```

```python
t, x = np.meshgrid(tt, xx)
```

```python
t.shape
```

```python
point = np.hstack([t.reshape(-1, 1), x.reshape(-1, 1)])
```

```python
point.shape
```

```python
point[-1, :]
```

```python
def u(X):
    if X.ndim == 2 and X.shape[1] != 1:
        x = X[:, 1]
        t = X[:, 0]
    elif X.ndim == 1:
        x = X
        
    k1, k2 = 0.4, 0.6
    a2 = 1./25
    x1, x2 = 4, 15
    theta1 = k1*x - k1**3 * t + x1
    theta2 = k2*x - k2**3 * t + x2
    up = 12*(k1 ** 2 * np.exp(theta1) + 
           k2 ** 2 * np.exp(theta2) + 
           2 * (k2 - k1) **2 * np.exp(theta1 + theta2)
           + a2 * (k2 ** 2 * np.exp(theta1) + k1**2 * np.exp(theta2)) * np.exp(theta1 + theta2))
    down = (1 + np.exp(theta1) + np.exp(theta2) + a2 * np.exp(theta1 + theta2)) ** 2
    u = up/down
    return u
```

```python
uu = u(point)
```

```python
exactu = uu.reshape(801, 1201)
```

```python
fig = plt.figure()
plt.plot(xx, exactu[:,0])
plt.xlabel("$x$")
plt.ylabel("$u$")
plt.show()
```

```python
T = 12
dt = T/len(tt)
fig = plt.figure()
plt.plot(xx, exactu[:, 400])
plt.xlabel("$x$")
plt.ylabel("$u$")
plt.show()
```

```python
readdata = scipy.io.loadmat("./data/kdvex4_2.mat")
```

```python
usol = readdata['uu']

# Grid
t_star = readdata['tt'][0]
x_star = readdata['x'][0]
TT, XX = np.meshgrid(t_star, x_star)

# Reference solution
plt.pcolor(TT, XX, usol, cmap='jet')
plt.colorbar()
plt.xlabel('$t$')
plt.ylabel('$x$')
```

```python
fig = plt.figure(figsize=(13, 4))
plt.subplot(1, 4, 1)
plt.plot(x_star, usol[:,0], color='blue')
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.title('$t = 0$')
plt.tight_layout()

plt.subplot(1, 4, 2)
plt.plot(x_star, usol[:,400], color='blue')
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.title('$t = 40$')
plt.tight_layout()

plt.subplot(1, 4, 3)
plt.plot(x_star, usol[:,800], color='blue')
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.title('$t = 80$')
plt.tight_layout()
plt.show()
fig = plt.figure(figsize=(13, 4))
plt.subplot(1, 4, 4)
plt.plot(x_star, usol[:,1200], color='blue')
plt.xlabel('$x$')
plt.ylabel('$u(t, x)$')
plt.title('$t = 120$')
plt.tight_layout()
plt.show()
```
